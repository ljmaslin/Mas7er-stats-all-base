```python
# from https://www.fxstreet.com/economic-calendar - needs updating frequently to catch Fed speaches
import pandas as pd
from google.colab import drive
import numpy
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/news.csv')
df['Start'] = pd.to_datetime(df['Start'])
df = df.set_index('Start')
df.sort_index(ascending=True)
strings = ["FOMC Minutes", "Fed's Chair", "Nonfarm Payrolls", "Consumer Price Index", "Producer Price Index", "Fed Interest Rate Decision"]
df = df[df['Name'].str.contains('|'.join(strings),regex=True)]
newsDays = numpy.unique(df.index.date)

years = [2015,2016,2017,2018,2019,2020,2021,2022,2023]

for y in years:
  newsDaysStr = []
  for nd in newsDays:
    if nd.year == y:
      newsDaysStr.append(f"(dom == {nd.day} and month == {nd.month} and year == {nd.year})")
  print(f"newsDay{y} = ", (' or ').join(newsDaysStr))
```
