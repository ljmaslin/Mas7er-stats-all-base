```python

for key, t in data_table.items():
  d = np.reshape(t[0], (-1, t[1])).T
  p_buckets = d[:,:10]
  t_buckets = d[:,10:]

  df1 = pd.DataFrame(p_buckets)
  #print(df1)
  df2 = pd.DataFrame(t_buckets)
  #print(df2)

  def get_day(i):
    day = "na"
    if i >= 0 and i <= 3:
      day = "mon"
    if i >= 4 and i <= 7:
      day = "tues"
    if i >= 8 and i <= 11:
      day = "weds"
    if i >= 12 and i <= 15:
      day = "thurs"
    if i >= 16 and i <= 19:
      day = "fri"

    if i == 0 or i == 1 or i == 4 or i == 5 or i == 8 or i == 9 or i == 12 or i == 13 or i == 16 or i == 17:
      day += "B1"
    else:
      day += "B2"

    if i % 2 != 0:
      day += "SHORT"
    else:
      day += "LONG"

    return day

  print("")
  print("")

  i = 0
  for _, row, in df1.iterrows():
    day = get_day(i)
    i+=1

    pc = f"var {key}_EXT{day} = array.from("
    for col in row:
      if col == "-":
        pc += "na,"
      else:
        pc+=f"{col},"
    pc = pc.rstrip(pc[-1])
    pc+=")"
    print(pc)

  i = 0
  for _, row, in df2.iterrows():
    day = get_day(i)
    i+=1

    pc = f"var {key}_TME{day} = array.from("
    for col in row:
      if "na" in col:
        pc += "na,"
      else:
        c = datetime.strptime(col, "%H:%M")
        pc+=f"{c.hour*3600*1000 + c.minute*60*1000},"
    pc = pc.rstrip(pc[-1])
    pc+=")"
    print(pc)
```
